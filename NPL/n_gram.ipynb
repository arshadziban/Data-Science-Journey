```python
import requests
from bs4 import BeautifulSoup

#Fetch the article
url = "https://realpython.com/python-pass/"
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

# Get the article content
# Real Python uses <div class="article-body"> for main content
content = soup.find("div", class_="article-body")
```

# Clean and Tokenize Text

```python
import re
```

```python
def fix_unicode(text):
    return text.replace(u"\u2019", "'")
```

```python
regex = r"[\w']+|[.]"
document = []
```

```python
if content:
    for paragraph in content.find_all("p"):
        cleaned = fix_unicode(paragraph.text)
        words = re.findall(regex, cleaned)
        document.extend(words)
else:
    print("Can not find the article")
```

```python
from collections import defaultdict
import random
```

```python
# Building Bigram Transitions
bigrams = zip(document, document[1:])
```

```python
bigrams
```

```python
# Creating the Transition Dictionary
transitions = defaultdict(list)

for prev, current in bigrams:
    transitions[prev].append(current)
```

```python
transitions
```

```python
# Sentence Generation Function
def generate_bigram_sentence():
    current = "."  # Start after a period
    result = []
    
    
    # Word Generation Loop
    while True:
        next_words = transitions.get(current)
        if not next_words:
            break
        
        # Selecting Next Word
        next_word = random.choice(next_words)
        result.append(next_word)
        current = next_word
        
        # Stopping Condition
        if current == ".":
            break
    return " ".join(result)
```

```python
for _ in range(5):
    print(generate_bigram_sentence())
```

```python
# Trigram Generator
```

```python
# Build trigram transitions
trigrams = zip(document, document[1:], document[2:])
trigram_transitions = defaultdict(list)
starts = []

for prev, current, nxt in trigrams:
    if prev == ".":
        starts.append(current)
    trigram_transitions[(prev, current)].append(nxt)
```

```python
def generate_trigram_sentence():
    prev = "."
    current = random.choice(starts)
    result = [current]

    while True:
        key = (prev, current)
        next_words = trigram_transitions.get(key)
        if not next_words:
            break
        next_word = random.choice(next_words)
        result.append(next_word)
        if next_word == ".":
            break
        prev, current = current, next_word
    return " ".join(result)
```

```python
generate_trigram_sentence()
```
